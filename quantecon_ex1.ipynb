## Exercise 1

```julia
function factorial2(n)
    res = 1
    for i in 1:n
        res = res * i
    end
    return res
end

factorial2(5)
```

## Exercise 2

```julia
function binomial_rv(n, p)
    k = 0
    ϵ = rand(n)
    for i in eachindex(ϵ)
        if ϵ[i] < p
            k = k + 1
        end
    end
    return k
end
```

```julia
binomial_rv(100, 0.5)
```

## Exercise 3

```julia
function myfun()
    n = 100000000 # Number of trials
    k = 0
    for i in 1:n
        k = k + (rand().^2 + rand().^2 <= 1)
    end
    return k/n * 4
end
```

```julia
n = 200
α = 0.9

function gendata(n, α)
    x = zeros(n)

    for i in 1:n-1
        x[i+1] = α*x[i] + randn()
    end
    return x
end


plot(1:n,gendata(n, α))
```

## Exercise 7

```julia
function fpt(a, tmax, α, σ)
    x = 1

    for i in 1:tmax
        x = α*x + σ*randn()
        if x <= a
            return i
        end
    end
    
    return tmax
end
```

```julia
n = 100000
t = zeros(n)
for i in 1:n
    t[i] = fpt(0.2, 200, 1.0, 0.2)
end
histogram(t, bins=20)
```

## Exercise 8(a)

```julia
function newton(f, f_prime, x_0, tol, maxiter)
    x_old = x_0
    
    for i in 1:maxiter
        x_new = x_old - f(x_old) / f_prime(x_old)
        if norm(x_old-x_new) < tol
            return 0.5 * (x_old + x_new)
        end
        x_old = x_new
    end
    
    println("Maximum number of iterations reached!")
    return 0.5* (x_old + x_new)
end
    
```
